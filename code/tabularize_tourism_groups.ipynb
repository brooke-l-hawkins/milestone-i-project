### Setup

```python
import pandas as pd
```

```python
# laod data
df = pd.read_csv('../data/derived/combined_data_by_year.csv', index_col=0)
```

```python
# select columns for tables
df = df[['year','tourism_group','tourism_rank_2012','country_txt_wb','arrivals','population','incident_ct_total', 'nvictim_total']]
```

### High tourism group

```python
# filter for high tourism group and 2012 data
high_2012_df = df[(df['tourism_group']=='high') & (df['year']==2012)]

# drop year and tourism group columns
high_2012_df = high_2012_df.drop(columns=['year','tourism_group'])

# examine result
high_2012_df
```

```python
# write to csv
high_2012_df.to_csv('../data/derived/high_tourism_2012_table.csv', index=False)
```

### Low tourism group

```python
# filter for low tourism group and 2012 data
low_2012_df = df[(df['tourism_group']=='low') & (df['year']==2012)]

# drop year and tourism group columns
low_2012_df = low_2012_df.drop(columns=['year','tourism_group'])

# examine result
low_2012_df
```

```python
# write to csv
low_2012_df.to_csv('../data/derived/low_tourism_2012_table.csv', index=False)
```
