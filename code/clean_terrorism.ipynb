# Clean dataset

### Filter by date

```python
# filter by date; events that occurred since 2012
df = df[df.iyear >= 2012]

# confirm filter worked as expected
print(f'Events occurred between {min(df.iyear)} and {max(df.iyear)}')
```

```python
# check dataset size
df.shape
```

### Filter for certainly terrorism

There is not always enough information available to determine whether an incident was definitively terrorism, "terrorism proper", rather than another form of crime (e.g. organized crime, hate crime, insurgency); see p. 11 and 16 of codebook. Here is the categorical breakdown for variable `doutterr`:

* 0 - No. There is essentially no doubt as to whether the incident is an act of terrorism.
* 1 - Yes. There is doubt as to whether the incident is an act of terrorism.

```python
df.doubtterr.value_counts()
```

```python
# choose whether to include only incidents where there is no doubt
remove_doubtful = True

if remove_doubtful:
    # filter for incidents that are definitively terrorism
    df = df[df.doubtterr == 0]
```

```python
# check dataset size
df.shape
```

### Filter for "successful" terrorism

A terrorist incident is considered successful if there are tangible effects. It does not necessarily mean that it achieved the perpetrator's original goal. If any attack type has a tangible effect, then it is considered "successful"; except for assassinations, which are only considered successful if the inteded target is killed. See p. 26 of codebook. Here is the categorical breakdown for variable `success`:

* 0 - No. The incident was not successful.
* 1 - Yes. The incident was successful.

```python
df.success.value_counts()
```

```python
# choose whether to include only "successful" incidents
remove_unsuccessful = True

if remove_unsuccessful:
    # filter for incidents that are "successful"
    df = df[df.success == 1]
```

```python
df.shape
```

### Potential filter: event type

Each incident may be one of multiple attack types, see p. 24 of codebook. Note that these events are assigned in a hierarchical order (different from variable order below) and at a maximum of three event types. For example, if one event included a hijacking, barricade, bombing, and unarmed assault, it would be recorded with `attacktype1`=2, `attacktype2`=4, `attacktype3`=5, and the unarmed assault would go unlisted. Here is the categorical breakdown for variable `attacktype1`, `attacktype2`, and `attacktype3`.

* 1 - Assassination
* 2 - Armed Assault
* 3 - Bombing/Explosion
* 4 - Hijacking
* 5 - Barricade Incident
* 6 - Kidnapping
* 7 - Facility/Infrastructure Attack
* 8 - Unarmed Assault
* 9 - Unknown

Here is a breakdown of events by type:

```python
pd.DataFrame({'attacktype1':df.attacktype1.value_counts(),
              'attacktype2':df.attacktype2.value_counts(),
              'attacktype3':df.attacktype3.value_counts()})
```

```python
# choose attack types to include
# include all attack types by setting to [1,2,3,4,5,6,7,8,9]
incident_types = [1,2,3,4,5,6,7,8,9]

# filter for incidents that include specified event type as primary, secondary or tertiary
df = df[df.attacktype1.isin(incident_types) |
        df.attacktype2.isin(incident_types) |
        df.attacktype3.isin(incident_types)]
```

```python
# check dataset size
df.shape
```

### Add columns

The total number of fatalities `nkill` "includes all victims *and* attackers who died as a result of the incident. If a figure is not reported or too vague, the field is blank. If only the perpetrator or victim fatalities are known, the field represents those numbers (so the number may be an underestimate). If multiple numbers are reported, the most recent value from a reputable resource will be reported. Conflicting reports of fatalities counts are noted in the Additional Notes `addnotes` field.

The total number of injured `nwound` "records the number of confirmed non-fatal injuries to both perpetrators and victims. It follows the same conventions as `nkill` (see above).

The suffix "us" in variables `nkillus` and `nwoundus` refer to fatalities and injuries (respectively) of US citizens, domestically or abroad.

The suffixes "ter" and "te" in variables `nkillter` and `nwoundte` refer to fatalities and injuries (respectively) of perpetartors.

See pp.50-51 of codebook for additional details.

I am adding a new field `nvictim` to total the number of fatalities and wounded, for all individuals (`nvictim`), US citizens (`nvictimus`), and perpetrators (`nvictimte`).

```python
# add victim count
df['nvictim']   = df['nkill']   +  df['nwound']
df['nvictimus'] = df['nkillus'] +  df['nwoundus']
df['nvictimte'] = df['nkillter'] + df['nwoundte']
```

```python
# confirm correct addition
df[['nkill','nwound','nvictim']][df.nkill>0].sample(5)
```

```python
# confirm correct addition
df[['nkillus','nwoundus','nvictimus']][df.nkillus>0].sample(5)
```

```python
# confirm correct addition
df[['nkillter','nwoundte','nvictimte']][df.nkillter>0].sample(5)
```

### Recast data types

I am recasting data types to match the Data Dictionary. Most data types imported correctly by default, with the exception of text fields, which imported as object types instead of string types. I correct that here.

```python
# find text columns
text_columns_indices = ['_txt' in c for c in df.columns]

# create dictionary of text column name and 'string'
text_columns = df.columns[text_columns_indices]
text_type = ['string' for c in text_columns]
text_convert_dict = dict(zip(text_columns, text_type))

# preview dictionary
text_convert_dict
```

```python
# convert text from object to string
df = df.astype(text_convert_dict)
```

```python
# check new data types to confirm successful conversion
df.dtypes
```

### Rename columns

I rename columns to match the Data Dictionary. This will make it easier to join datasets later on.

```python
# rename columns to match data dictionary
df.rename(columns={'iyear':'year_gtd', 'imonth':'month_gtd', 'iday':'day_gtd', 
                   'country':'country_id', 'country_txt':'country_txt_gtd',
                   'nkillter':'nkillte'}, inplace=True)
```

```python
# check new column names to confirm successful renaming
for c in df.columns:
    print(c)
```

### Reset index

I've filtered away many records, and will now reset the index for a fresh start on the newly cleaned dataset.

```python
df.reset_index(drop=True, inplace=True)
```
