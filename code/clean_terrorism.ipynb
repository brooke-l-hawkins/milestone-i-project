# Clean dataset

### Filter by date

```python
# filter by date; events that occurred since 2012
df = df[df.iyear >= 2012]

# confirm filter worked as expected
print(f'Events occurred between {min(df.iyear)} and {max(df.iyear)}')
```

```python
# check dataset size
df.shape
```

### Filter by event type

Each incident may be one of multiple attack types, see p. 24 of codebook. Note that these events are assigned in a hierarchical order (different from variable order below) and at a maximum of three event types. For example, if one event included a hijacking, barricade, bombing, and unarmed assault, it would be recorded with `attacktype1`=2, `attacktype2`=4, `attacktype3`=5, and the unarmed assault would go unlisted. Here is the categorical breakdown for variable `attacktype1`, `attacktype2`, and `attacktype3`.

* 1 - Assassination
* 2 - Armed Assault
* 3 - Bombing/Explosion
* 4 - Hijacking
* 5 - Barricade Incident
* 6 - Kidnapping
* 7 - Facility/Infrastructure Attack
* 8 - Unarmed Assault
* 9 - Unknown

Here is a breakdown of events by type:

```python
pd.DataFrame({'attacktype1':df.attacktype1.value_counts(),
              'attacktype2':df.attacktype2.value_counts(),
              'attacktype3':df.attacktype3.value_counts()})
```

```python
# choose attack types to include
# include all attack types by setting to [1,2,3,4,5,6,7,8,9]
incident_types = [1,2,3,4,5,6,7,8,9]

# filter for incidents that include specified event type as primary, secondary or tertiary
df = df[df.attacktype1.isin(incident_types) |
        df.attacktype2.isin(incident_types) |
        df.attacktype3.isin(incident_types)]
```

```python
# check dataset size
df.shape
```

### Filter for certainly terrorism

There is not always enough information available to determine whether an incident was definitively terrorism, "terrorism proper", rather than another form of crime (e.g. organized crime, hate crime, insurgency); see p. 11 and 16 of codebook. Here is the categorical breakdown for variable `doutterr`:

* 0 - No. There is essentially no doubt as to whether the incident is an act of terrorism.
* 1 - Yes. There is doubt as to whether the incident is an act of terrorism.

```python
df.doubtterr.value_counts()
```

```python
# choose whether to include only incidents where there is no doubt
remove_doubtful = False

if remove_doubtful:
    # filter for incidents that are definitively terrorism
    df = df[df.doubtterr == 0]
```

```python
# check dataset size
df.shape
```

### Filter for "successful" terrorism

A terrorist incident is considered successful if there are tangible effects. It does not necessarily mean that it achieved the perpetrator's original goal. If any attack type has a tangible effect, then it is considered "successful"; except for assassinations, which are only considered successful if the inteded target is killed. See p. 26 of codebook. Here is the categorical breakdown for variable `success`:

* 0 - No. The incident was not successful.
* 1 - Yes. The incident was successful.

```python
df.success.value_counts()
```

```python
# choose whether to include only "successful" incidents
remove_unsuccessful = False

if remove_unsuccessful:
    # filter for incidents that are "successful"
    df = df[df.success == 1]
```

```python
df.shape
```

# Group by variables

### Collapse related incidents

If an event occurs at separate times or locations, than they are considered separate incidents; see p. 12 of codebook. "If any discontinuity in time or space is noted, the event is comprised of multiple incidents."

Relevant variables are `multiple`, which indicates whether attacks are connected; see p. 17 of codebook. If the attack is part of a multiple incident, the variable `related` includes the text of event ID's from `eventid` for related incidents; see p. 13 and 17 of codebook.

```python
df.shape
```

Collapse events into one incident if records are related, occurred in the same city and on the same day. Numbers of victims are summed across records into a single incident.

```python
# group related incidents that occur on same city and day
terrorism_by_incident = df.groupby(['related',
                                    'iyear','imonth','iday',
                                    'country','region',
                                    'country_txt','region_txt',
                                    'provstate','city']).agg(
    incident_id  = ('eventid', 'first'),
    nkill     = ('nkill', 'sum'),
    nkillus   = ('nkillus', 'sum'),
    nkillter  = ('nkillter', 'sum'),
    nwound    = ('nwound', 'sum'),
    nwoundus  = ('nwoundus', 'sum'),
    nwoundte  = ('nwoundte', 'sum'),
    event_ct_total   = ('eventid', 'size')
)

# reset index
terrorism_by_incident.reset_index(inplace=True)
terrorism_by_incident.set_index('incident_id', inplace=True)

# rename columns
terrorism_by_incident = terrorism_by_incident.rename(columns={'iyear':'year_gtd',
                                                              'imonth':'month_gtd',
                                                              'iday':'day_gtd',
                                                              'country':'country_gtd',
                                                              'country_txt':'country_txt_gtd'})

# inspect dataframe
terrorism_by_incident.head()
```

```python
# write to csv
terrorism_by_incident.to_csv('../data/derived/terrorism_by_incident.csv')
```

### Group by year

```python
# group by year and country
terrorism_by_year = terrorism_by_incident.groupby(['year_gtd','country_gtd']).agg(
    incident_ct_total = ('country_gtd', 'size'), # arbitrarily chose to count one variable
    nkill_total       = ('nkill', 'sum'),
    nkillus_total     = ('nkillus', 'sum'),
    nkillter_total    = ('nkillter', 'sum'),
    nwound_total      = ('nwound', 'sum'),
    nwoundus_total    = ('nwoundus', 'sum'),
    nwoundte_total    = ('nwoundte', 'sum')
)

# inspect dataframe
terrorism_by_year.head()
```

```python
# write to csv
terrorism_by_year.to_csv('../data/derived/terrorism_by_year.csv')
```
