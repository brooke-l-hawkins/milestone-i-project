### Filter by date

### Filter by event type

Each incident may be one of multiple attack types, see p. 24 of codebook. Note that these events are assigned in a hierarchical order and at a maximum of three event types. For example, if one event included a hijacking, barricade, bombing, and unarmed assault, it would be recorded with attacktype1=2, attacktype2=4, attacktype3=5, and the unarmed assault would go unlisted. Here is the categorical breakdown for variable "attacktype1", "attacktype2", and "attacktype3".

1. Assassination
2. Hijacking
3. Kidnapping
4. Barricade Incident
5. Bombing/Explosion
6. Armed Assault
7. Unarmed Assault
8. Facility/Infrastructure Attack
9. Unknown

@Jim - in my opinion, we may not care to filter based on this variable. What I usually think of when I think of terrorism and travel is 5 (bombing) and 8 (facility/infrastructure attack), but any of these event types could have an impact on tourist interest, because they're all scary.

Here is a breakdown of events by type:

```python
df.attacktype1.value_counts()
```

```python
df.attacktype2.value_counts()
```

```python
df.attacktype3.value_counts()
```

```python
# choose attack types to include
# include all attack types by setting to [1,2,3,4,5,6,7,8,9]
incident_types = [1,2,3,4,5,6,7,8,9]

# filter for incidents that include specified event type as primary, secondary or tertiary
df = df[df.attacktype1.isin(incident_types) |
        df.attacktype2.isin(incident_types) |
        df.attacktype3.isin(incident_types)]
```

```python
# check dataset size
df.shape
```

There is not always enough information available to determine whether an incident was definitively terrorism, "terrorosim proper", rather than another form of crime (e.g. organized crime, hate crime, insurgency); see p. 11 and 16 of codebook. Here is the categorical breakdown for variable "doutterr":

0. No. There is essentially no doubt as to whether the incident is an act of terrorism.
1. Yes. There is doubt as to whether the incident is an act of terrorism.

```python
# choose whether to include only incidents where there is no doubt
remove_doubtful = False

if remove_doubtful:
    # filter for incidents that are definitively terrorism
    df = df[df.doubtterr == 0]
```

### Filter for "successful" terrorism

A terrorist incident is considered successful if there are tangible effects. It does not necessarily mean that it achieved the perpetrator's original goal. If any attack type has a tangible effect, then it is considered "successful"; except for assassinations, which are only considered successful if the inteded target is killed. See p. 26 of codebook. Here is the categorical breakdown for variable "success":

1. Yes. The incident was successful.
2. No. The incident was not successful.

@Me - check out whether this variable has a big effect; are there many unsuccessful incidents? do they have any casualties/wounded/etc?

```python
df.success.value_counts()
```

```python
# choose whether to include only "successful" incidents
remove_unsuccessful = True

if remove_unsuccessful:
    # filter for incidents that are "successful"
    df = df[df.success == 1]
```
