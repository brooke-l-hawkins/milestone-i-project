### Filter by event type

Each incident may be one of multiple attack types, see p. 24 of codebook. Note that these events are assigned in a hierarchical order and at a maximum of three event types. For example, if one event included a hijacking, barricade, bombing, and unarmed assault, it would be recorded with attacktype1=2, attacktype2=4, attacktype3=5, and the unarmed assault would go unlisted. Here is the categorical breakdown for variable "attacktype1", "attacktype2", and "attacktype3".

1. Assassination
2. Hijacking
3. Kidnapping
4. Barricade Incident
5. Bombing/Explosion
6. Armed Assault
7. Unarmed Assault
8. Facility/Infrastructure Attack
9. Unknown

*Although we originally planned to filter on this variable, in my opinion, we probably don't need to. What I usually think of when I think of terrorism and travel is 5 (bombing) and 8 (facility/infrastructure attack), but any of these event types could have an impact on tourist interest, because they're all scary.*

Here is a breakdown of events by type:

```python
# choose attack types to include
# include all attack types by setting to [1,2,3,4,5,6,7,8,9]
incident_types = [1,2,3,4,5,6,7,8,9]

# filter for incidents that include specified event type as primary, secondary or tertiary
df = df[df.attacktype1.isin(incident_types) |
        df.attacktype2.isin(incident_types) |
        df.attacktype3.isin(incident_types)]
```

### Filter for "successful" terrorism

A terrorist incident is considered successful if there are tangible effects. It does not necessarily mean that it achieved the perpetrator's original goal. If any attack type has a tangible effect, then it is considered "successful"; except for assassinations, which are only considered successful if the inteded target is killed. See p. 26 of codebook. Here is the categorical breakdown for variable "success":

1. Yes. The incident was successful.
2. No. The incident was not successful.

*I think this will have an effect on our visualizations. My guess is that unsuccessful incidents (usually) have no wounded or killed. Thinking about our violin plots, including unsuccessful events may lead to a distribution that is a little bottom heavy. Just something to keep in mind!*

```python
df.success.value_counts()
```

```python
# choose whether to include only "successful" incidents
remove_unsuccessful = False

if remove_unsuccessful:
    # filter for incidents that are "successful"
    df = df[df.success == 1]
```

### Collapse related incidents

If an event occurs at separate times or locations, than they are considered separate incidents; see p. 12 of codebook. "If any discontinuity in time or space is noted, the event is comprised of multiple incidents."

Relevant variables are "multiple", which indicates whether attacks are connected; see p. 17 of codebook. Here is the categorical breakdown for the variable multiple:

1. Yes. The attack is part of a multiple incident.
2. No. The attack is not part of a multiple incident.

If the attack is part of a multiple incident, the variable related "related" includes the text of event ID's from "eventid" for related incidents; see p. 13 and 17 of codebook.

```python
df.shape
```

*The data dictionary includes many more columns, but to keep them and to aggregate related events will require some aggregation function for those other columns. If we want more columns, let's talk about the best way to aggregate them in this step!*

Here's an example for how to concatenate text; similar approach could be used for transforming other columns as well.

```python
#df.groupby(['related','iyear','imonth','iday',
#            'country','country_txt','region',
#            'region_txt','provstate','city'])['summary'].apply(','.join)
```

Collapse events into one incident if records are related, occurred in the same city and on the same day. Numbers of victims are summed across records into a single incident.

```python
# group related incidents that occur on same city and day
terrorism_by_incident = df.groupby(['related',
                                    'iyear','imonth','iday',
                                    'country','region',
                                    'country_txt','region_txt',
                                    'provstate','city']).agg(
    incident_id  = ('eventid', 'first'),
    nevents   = ('eventid', 'size'),
    doubtterr = ('doubtterr', 'sum'),
    success   = ('success', 'sum'),
    targtour  = ('targtour', 'sum'),
    nkill     = ('nkill', 'sum'),
    nkillus   = ('nkillus', 'sum'),
    nkillter  = ('nkillter', 'sum'),
    nwound    = ('nwound', 'sum'),
    nwoundus  = ('nwoundus', 'sum'),
    nwoundte  = ('nwoundte', 'sum')
)

# reset index
terrorism_by_incident.reset_index(inplace=True)
terrorism_by_incident.set_index('incident_id', inplace=True)

# rename columns
terrorism_by_incident = terrorism_by_incident.rename(columns={'iyear':'year_gtd',
                                                              'imonth':'month_gtd',
                                                              'iday':'day_gtd',
                                                              'country':'country_gtd',
                                                              'country_txt':'country_txt_gtd'})

# inspect dataframe
terrorism_by_incident.head()
```

```python
# write to csv
terrorism_by_incident.to_csv('../data/derived/terrorism_by_incident.csv')
```

### Group by year

```python
# group by year and country
terrorism_by_year = terrorism_by_incident.groupby(['year_gtd','country_gtd']).agg(
    incident_ct_total = ('country_gtd', 'size'), # arbitrarily chose to count one variable
    nkill_total       = ('nkill', 'sum'),
    nkillus_total     = ('nkillus', 'sum'),
    nkillter_total    = ('nkillter', 'sum'),
    nwound_total      = ('nwound', 'sum'),
    nwoundus_total    = ('nwoundus', 'sum'),
    nwoundte_total    = ('nwoundte', 'sum')
)

# inspect dataframe
terrorism_by_year.head()
```

```python
# write to csv
terrorism_by_year.to_csv('../data/derived/terrorism_by_year.csv')
```
